```python
df=data.copy()
```

```python
#Spliting Input column and the output column
x=df.iloc[:,0:4]
y=df.iloc[:,4]
```

```python
# Separating the data into tranning and testing
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

MODEL-1:Support vector machine algorithm

```python
#At first we train our sample data from dataset x_train and y_train
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
model_svc=SVC()
model_svc.fit(x_train,y_train)
```

```python
prediction_train=model_svc.predict(x_train)
print("Training Accuracy=",accuracy_score(y_train,prediction_train)*100)
```

```python
prediction1=model_svc.predict(x_test)
print("Test Accuracy=",accuracy_score(y_test,prediction1)*100)
```

MODEL-2: Logistic Regression Algorithm

```python
from sklearn.linear_model import LogisticRegression
model_lregression=LogisticRegression()
model_lregression.fit(x_train,y_train)
```

```python
prediction_train=model_lregression.predict(x_train)
print("Training Accuracy=",accuracy_score(y_train,prediction_train)*100)
```

```python
prediction2=model_lregression.predict(x_test)
print("Test Accuracy=",accuracy_score(y_test,prediction2)*100)
```

```python
#A details classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,prediction2))
```

For precision value-1(Iris-setosa & Iris-versicolor) means all actual postive categories are 100% positive predicted outcomes.There are no negative or incorrectly categories outcomes.

For precision value-0.86(Iris-virginica) means there are some peredicted values incorrectly outcomes.
